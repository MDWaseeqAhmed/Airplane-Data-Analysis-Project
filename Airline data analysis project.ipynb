# Analysis

### Question 1: How many planes have more than 100 seats?

```python
pd.read_sql_query("""select aircraft_code, count(*) as num_seats from seats group by aircraft_code
                            having num_seats > 100""", connection)
```

### Question 2: How the number of tickets booked and total amount earned changed with time?

```python
tickets = pd.read_sql_query("""select * from tickets inner join bookings on 
                                    tickets.book_ref = bookings.book_ref""", connection)

tickets['book_date'] = pd.to_datetime(tickets['book_date'])
tickets['date'] = tickets['book_date'].dt.date
tickets
```

```python
line_graph = tickets.groupby('date')[['date']].count()
plt.figure(figsize = (18,6))
plt.plot(line_graph.index, line_graph['date'], marker = '^')
plt.xlabel('Date', fontsize =20)
plt.ylabel('Number of Tickets', fontsize = 20)
plt.grid()
plt.show()
```

```python
booking = pd.read_sql_query("""select * from bookings """, connection)
booking['book_date'] = pd.to_datetime(booking['book_date'])
booking['date'] = booking['book_date'].dt.date
booking
```

```python
booking = booking.groupby('date')[['total_amount']].sum()
plt.figure(figsize = (18,6))
plt.bar(booking.index, booking['total_amount'], color ='maroon')
plt.xlabel('Date', fontsize =20)
plt.ylabel('Total Amount', fontsize = 20)
plt.grid()
plt.show()
```

# Analyzing Occupancy Rate

### Question 4: For each aircraft, calculate the total revenue per year and the average return per ticket?

```python
pd.read_sql_query("""select aircraft_code, ticket_count, total_revenue, total_revenue/ticket_count as avg_revenue from 
(select aircraft_code, count(*) as ticket_count, sum(amount) as total_revenue from ticket_flights
                    join flights on ticket_flights.flight_id = flights.flight_id group by aircraft_code)""", connection)
```

### Question 5: Calculate the average occupancy per aircraft?

```python
occupancy_rate = pd.read_sql_query("""select a.aircraft_code, avg(a.seats_count) as booked_seats, 
        b.num_seats, avg(a.seats_count)/b.num_seats as occupancy_rate from
            (select aircraft_code, flights.flight_id, count(*) as seats_count from boarding_passes
            inner join flights on boarding_passes.flight_id=flights.flight_id 
            group by aircraft_code, flights.flight_id) as a
            inner join
            (select aircraft_code, count(*) as num_seats from seats group by aircraft_code) as b
            on a.aircraft_code = b.aircraft_code group by b.aircraft_code""", connection)

occupancy_rate
```

### Question 6: Calculate by how much the total annual turnover could increase by giving all airraft a 10% higher occupancy rate?

```python
occupancy_rate['Inc occupancy rate'] = occupancy_rate['occupancy_rate']+occupancy_rate['occupancy_rate']*0.1

occupancy_rate
```

```python
total_revenue = pd.read_sql_query("""select aircraft_code, sum(amount) as total_revenue from ticket_flights
            join flights on ticket_flights.flight_id = flights.flight_id group by aircraft_code""", connection)
total_revenue
```

```python
occupancy_rate['Inc Total Annual Turnover'] = (total_revenue['total_revenue']/
                                               occupancy_rate['occupancy_rate']*occupancy_rate['Inc occupancy rate'])
occupancy_rate
```
